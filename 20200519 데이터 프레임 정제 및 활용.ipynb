## 행 1개로 구성된 데이터 프레임 생성하여 연결하기

```python
new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns=['A','B','C','D'])
new_row_df
```

```python
pd.concat([df1, new_row_df])
#칼람 명을 정학히 주고, 데이터를 concat하면 columns 네임에 맞게 들어가진다.
# concat 메서드는 한번에 2개 이상의 데이터프레임도 연결할 수 있다.
```

## 공통 열과 인덱스만 연결하기 

### 열로 연결하기 

```python
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
pd.concat([df1,df2,df3])
# 공콩 컬럼이 부족하기 때문에 누락값 발생할 수 있다. 
```

```python
pd.concat([df1,df2,df3], join = 'inner')
# 공통 열만 골라서 연결하기 위해선 join인자를 inner로 지정한다
#empty로 나오는 이유는 3개의 데이터프레임이 모두 가지고 있는 열이 없기 때문
```

```python
pd.concat([df1,df3], ignore_index=False, join = 'inner')
#df1과 df3는 A,C라는 공통 되는 열이 있기 때문에 join = 'inner' 조건이 적절하게 발동한다. 
#ignore_index 를 True 두면, 데이터프레임을 concat할때 기존의 index를 무시하고 순서대로 다시 생긴다. 
```

### 행으로 연결하기

```python
df1.index = [0,1,2,3]
df2.index = [4,5,6,7]
df3.index = [0,2,5,7]
```

```python
pd.concat([df1,df2,df3], axis = 1)
#axis =1 옵션은 데이터프레임들을 열로 붙인다는것, 세로가 아닌 가로! 옵션을 없애주면 세로로 행으로 붙인다.
```

```python
pd.concat([df1,df3], axis = 1, join = 'inner')
#df1, df3는 공통 행을 가지고 있기 떄문에, axis = 1, join='inner'는 행기준으로 같은 인덱스 끼리 join 하라는 뜻이다. 
```

# merge 메서드 사용하기
## 연결 전용 메서드

```python
person = pd.read_csv('../data/survey_person.csv')
site = pd.read_csv('../data/survey_site.csv')
survey = pd.read_csv('../data/survey_survey.csv')
visited =  pd.read_csv('../data/survey_visited.csv')
```

```python
visited_subset = visited.loc[[0,2,6],]
print(visited_subset)
print(site)
```

```python
# merge 기본적으로 내부 조인을 실행하며, 메서드를 사용한 데이터 프레임을 왼쪽으로 지정, 
# 첫 번째 인자값으로 지정한 데이터 프레임을 오른쪽으로 지정, 열이름 일치 - 왼쪽 프레임을 기준으로 연결
new_merge = site.merge(visited_subset, left_on = 'name', right_on='site')
new_merge
```

```python
ps = person.merge(survey, left_on='ident', right_on='person')
#left, right_on으로 들어가는 컬럼들은 공통되는 데이터를 가진 컬럼의 열이름이어야 한다. left는 메서드를 사용한 데이터 프레임의 열,
#right에는 인자값으로 지정한 데이터 프레임의 열이다. 
ps.head()
```

```python
vs = visited.merge(survey, left_on= 'ident', right_on='taken')
vs.head()
```

```python
# ps 데이터 프레임의 ident, taken, quant, reading 열의 값과 vs데이터 프레임의 person.
#ident, quant, reading 열의 값을 이용하여 ps와 vs 데이터 프레임을 연결 
ps_vs = ps.merge(vs, left_on = ['ident','taken','quant','reading'],  right_on = ['person', 'ident', 'quant', 'reading'])
#left, right_on 들로 merge 되는 칼럼들은 같은 데이터를 가진 칼럼들이다. 
ps_vs.head()
```

# 누락값 처리하기 - 변경,삭제
## 1. 누락값 변경하기 
* 0으로 채워버리기

```python
ebola.fillna(0).iloc[0:10, 0:5]
#10행 5열 까지 누락값 NaN을 0으로 채우겠다-> fillna(0)
```

* 전의 값으로 누락값 변경

```python
ebola.fillna(method='ffill').iloc[0:10, 0:5]
#10행 5열까지 결측치가 확인되면, 그 앞의값(윗 행의 값)을 가져온다- -> fillna(method='fjill')
```

* 후의 값으로 누락값 변경

```python
ebola.fillna(method='bfill').iloc[0:10, 0:5]
#10행 5열까지 결측치가 확인되면, 그 뒤의값(아랫 행의 값)을 가져온다- -> fillna(method='bjill')
```

* 누락값 위아래로 있는 값의 중간값으로 채우기

```python
ebola.interpolate().iloc[0:10, 0:5]
```

# 누락값이 포함된 데이터 계산하기

```python
ebola['Cases_multiple'] = ebola[['Cases_Guinea', 'Cases_Liberia','Cases_SierraLeone']].sum(axis=1)
#axis =1 옵션은 데이터프레임들을 열로 붙인다는것, 지정된 열들의 값을 sum하고, Cases_multiple 이라는 열이름으로 데이터를 삽입
ebola.tail()
print(ebola.Cases_multiple.sum())
```
