## 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

## 1. 히스토그램은 df의 열데이터 분포와 빈도를 살펴보는 용도로 자주 사용됨


```python
tips = sns.load_dataset("tips")
print(tips.head(4))
```

### 기본틀 생성하기

```python
#fig(기본틀), 그래프 격자(axes1)
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

### 변수를 하나만 사용해서 그린 일변량 그래프

```python
axes1.hist(tips['total_bill'], bins = 10)#bins=10 : x축 간격을 10으로 조정 
axes1.set_title('hist of total bill')#제목 설정
axes1.set_xlabel('Freguency')#x축 설정
axes1.set_ylabel('Total Bill')#y축 설정
fig
```

## 2. 산점도
### 변수 2개 사용한 이변량 그래프->total bill 열에 따른 tip열의 분포를 나타낸 산점도 그래프 

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])#산점도 그래프-> 변수를 2개 사용하고, scatter 함수 사용. scatter(x축데이터, y축데이터)
axes1.set_title('Scatterplot of bill vs tip')
axes1.set_xlabel('totalbill')
axes1.set_ylabel('tip')
```

## 3. 박스 그래프 
### 이산형 변수-female, male과 같이 명확히 구분되는 값
### 연속형 변수 -Tip과 같이 명확히 셀수 없는 변위의 값

```python
boxplot = plt.figure()#틀생성
boxplot.suptitle('Tips')
axes1= boxplot.add_subplot(1,2,1)#1행2열 첫번쩨 플롯 생성
axes2 = boxplot.add_subplot(1,2,2)#1행2열 두번째 한개의 플롯 생성
axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],#성별이 Female 사람들의 tip
     tips[tips['sex'] == 'Male']['tip']],#성별이 male 사람들의 tip
     labels=['Female', 'Male'])

axes2.boxplot(
    [tips[tips['time'] == 'Dinner']['tip'],#시간이 Dinner일떄의  tip
     tips[tips['time'] == 'Lunch']['tip']],#시간이 Lunch 사람들의 tip
     labels=['Dinner', 'Lunch'],)
```

## 다변량 데이터로 다변량 그래프 그리기 - 산점도 
### 3개이상의 다변량데이터로 그래프그리기

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else: return 1
tips
    
tips['sex_color'] = tips['sex'].apply(recode_sex) 
#recod_sex 라는 함수에, tips['sex']의 데이터들을 apply(적용)함으로서 매개변수를 넘겨주고, 그 결과를 sex_color에 저장한다. 
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x= tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] *30, # 산점도의 점 크기를 size데이터의 크기로 잡는다
    c = tips['sex_color'], # 컬러를 sex_color에 맞춘다. 
    alpha = 0.5)
axes1.set_title('total bill vs tip colored by sex and sized by size')
axes1.set_xlabel('total bill')
axes1.set_ylabel('Tips')
```
