```python
import matplotlib as plt
%matplotlib inline
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
#f_path = "/Library/Fonts/AppleGothic.ttf (맥, Mac)
f_path = "C:/Windows/Fonts/malgun.ttf"
#(윈도우즈,Windows)
font_name = font_manager.FontProperties(fname= f_path).get_name()
rc('font', family = font_name)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('../data/nexercise.xls')
data.head()
```

```python
data1 = data[data['대분류'] == '성별']
data1
```

```python
data2 = data[data['대분류'] == '연령별']
data2
```

```python
f, ax = plt.subplots(1,2, figsize = (13,6))
f1, ax1 = plt.subplots(1,3, figsize = (20,6))

data1['운동을 할 충분한 시간이 없어서'].plot.pie(ax = ax[0], autopct = '%1.1f%%',
                                    labels= data1['분류'],
                           fontsize = 20, explode = (0.02, 0))
ax[0].set_title('운동을 할 충분한 시간이 없어서', size =  20)
ax[0].set_ylabel('')

data1['함께 운동을 할 사람이 없어서'].plot.pie(ax = ax[1], autopct = '%1.1f%%',
                                    labels= data1['분류'],
                           fontsize = 20, explode = (0.02, 0))
ax[1].set_title('함께 운동을 할 사람이 없어서', size =  20)
ax[1].set_ylabel('')

explode_setting = [0.02, 0.02, 0.02, 0.02, 0.02, 0.02]
data2['운동을 할 충분한 시간이 없어서'].plot.pie(ax = ax1[0], autopct = '%1.1f%%', 
                               labels= data2['분류'],
                          startangle= 90, fontsize = 20, explode = explode_setting)
ax1[0].set_title('운동을 할 충분한 시간이 없어서', size =  20)
ax1[0].set_ylabel('')

data2['함께 운동을 할 사람이 없어서'].plot.pie(ax = ax1[1], autopct = '%1.1f%%', 
                               labels= data2['분류'],
                          startangle= 90, fontsize = 20, explode = explode_setting)
ax1[1].set_title('함께 운동을 할 사람이 없어서', size =  20)
ax1[1].set_ylabel('')

data2['운동을 할 만한 장소가 없어서'].plot.pie(ax = ax1[2], autopct = '%1.1f%%', 
                             labels= data2['분류'],
                          startangle= 90, fontsize = 20, explode = explode_setting)
ax1[2].set_title('운동을 할 만한 장소가 없어서', size =  20)
ax1[2].set_ylabel('')
```
