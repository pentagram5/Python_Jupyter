```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
seoul_data = pd.read_excel("../data/seoul.xls")
seoul_data.head(2)
```

```python
seoul_age = seoul_data[seoul_data['대분류']== '연령별']
seoul_age
```

```python
f, ax = plt.subplots(1,3, figsize = (20,6))
explode_setting = [0.02, 0.02, 0.02, 0.02, 0.02, 0.02]
#f, ax로 subplots를 설정하면 figure과 axes 속성을 하나하나 지정할 수 있다. (1,3)으로 지정했기 떄문에 ax[0]~ax[2]까지 표현된다.

seoul_age['TV  또는 비디오시청'].plot.pie(ax = ax[0], autopct = '%1.1f%%', labels= seoul_age['분류'],
                          startangle= 90, fontsize = 20, explode = explode_setting)
ax[0].set_title('연령별 TV  또는 비디오시청 통계', size =  20)
ax[0].set_ylabel('')

seoul_age['종교활동'].plot.pie(ax = ax[1], autopct = '%1.1f%%', labels= seoul_age['분류'],
                          startangle= 90,fontsize = 20,explode = explode_setting)
ax[1].set_title('연령별 종교 활동 통계', size =  20)
ax[1].set_ylabel('')

seoul_age['문화예술관람'].plot.pie(ax = ax[2], autopct = '%1.1f%%', labels= seoul_age['분류'],
                          startangle= 90, fontsize = 20, explode = explode_setting)
ax[2].set_title('연령별 문화예술 관람 통계', size =  20)
ax[2].set_ylabel('')
```
