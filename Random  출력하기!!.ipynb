```python
import random
#random 함수를 쓰기 위해 패키지 호출

num_list_tot = []
#랜덤 출력한 숫자 조합들을 저장할 리스트

for i in range(0,500):
#while 반복문을 사용하면 커널에 오버헤드가 발생하여 굉장한 버퍼를 일으키므로, for 문의 조건절에 충분히 큰 숫자를 사용
# 조합의 경우의 수를 모를때는 이렇게 하는 수 밖에 읍따!

    num_list = []
    #랜덤으로 추출될 수를 저장할 리스트 
    
    list_1 = [6,7,9,12,15,16,17,25,45]
    #매 반복시다마, list_1에서 숫자를 pop하기 때문에 숫자 조합이 완료되면 다시 초기화를 해주어야 한다.
    
    for i in range(0,6): # 6번 반복
        r_seed = random.randrange(0,len(list_1)) 
        #0부터 list_1의 길이만큼 의 수중, 랜덤으로 하나를 뽑는다.
        get_num = list_1.pop(r_seed) # num_list에서 숫자 뽑은 뒤 저장
        #뽑아진 숫자를 list_1.pop에 넘겨준다 ex) 1이 랜덤출력되면, list_1.pop(1)= 7을 넘겨준다. pop이기 때문에 list_1에선 제거.
        num_list.append(get_num)
        #numlist에 list_1에서 pop된 숫자를 append(추가)해준다.
        num_list.sort()
        #들어온 숫자들을 정렬한다.
    if num_list_tot.count(num_list) == 1 :
        pass
    # 만약 조합들중 하나라도 전체조합을 넣는 리스트에 중복된다면, pass 아래의 조건문은 수행하지 않는다.
    
    else: num_list_tot.append(num_list)
    # 조합중 중복되는게 없다면, num_list_tot에 숫자조합 num_list를 넘겨준다. 
    
num_list_tot.sort()
#조합들을 다시 정렬-> 오름차순
count = 1
for i in num_list_tot:
    print("%2d. -- >%s"%(count, i), end = '\n')
    count +=1
    #num_list_tot 내용들을 출력해준다.
    
print("\n조합의 총 개수는 %d이다"%(len(num_list_tot)))
#num_list_tot의 길이를 출력한다. 길이 = 갯수
rannum = random.randrange(0,len(num_list_tot))
print("\n조합 중, 랜덤 출력한 결과는....... %d번째의 조합 %s 이다!!!"%(rannum, num_list_tot[rannum]))

```
