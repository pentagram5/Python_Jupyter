```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python

#밀집도(정규화 시켜 넓이가 1이 되도록 그린 그래프) 제외
# kernel Density Estimation(커널밀도추정)   kde
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)   
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)   
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
#rug인자 추가 - 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프
hist_den_rug, ax =  plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)   
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python

#count 그래프 - 이산값을 나타낸 그래프
ax =  plt.subplots()
ax = sns.countplot('day', data = tips)
ax.set_title('Count of Days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
#다양한 종류의 이변량 그래프 그리기
# seaborn 라이브러리로 산점도 그래프 그리기
# regplot메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다.
ax =  plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python

joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```python
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```python
# 이차원 밀집도 그리기
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                data2 = tips['tip'],
                shade = True)      # shade = True는 음영 효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python

# 이차원 밀집도 그리기
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                data2 = tips['tip'],
                shade = False)      # shade = True는 음영 효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#관계 그래프 그리기
fig = sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips) #대각선 방향 왼쪽위에서 오른쪽 아래 
pair_grid = pair_grid.map_upper(sns.regplot)# 대각선 기준 위쪽 -regplot메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다.
pair_grid = pair_grid.map_lower(sns.kdeplot)# 대각선 기준 아랫쪽 - kdeplot - 이차원 밀집도 그리기
pair_grid = pair_grid.map_diag(sns.distplot, rug = True) # 대각선 중심 
plt.show()
```

### 산점도 그래프의 크기와 모양 조절하기 

```python
#산점도 그래프의 점 크기 : scatter_kws에 딕셔너리 형태로 인자값을 전달 
scatter = sns.lmplot(x = 'total_bill', y='tip', data = tips, fit_reg = False, hue = 'sex', scatter_kws={'s':tips['size']*10})
plt.show()
```

### 그래프에 스타일 적용해보기

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips, split = True)
```

```python
sns.set_style('whitegrid')# 격자추가하기 
fig, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips, split = True)
```

```python
fig = plt.figure()
seaborn_style = ['darkgrid','whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_style):
#enumerate (리스트) --> [(0,리스트[0]), (1,리스트[1], ....., (n, 리스트[n])] : 열거하다라는 뜻을 가짐쓰
# 따라서, 위의 for문에서 idx는 리스트의 인덱스를 가르키고, style은 리스트의 내용을 가리킨다. 
    plot_position = idx +1
    with sns.axes_style(style):
        ax = fig.add_subplot(2,3, plot_position)
        violin = sns.violinplot(x = 'time', y = 'total_bill', data = tips, ax = ax)
        violin.set_title(style)
fig.tight_layout()
```

## matplotlib.pyplot 활용해보기 

```python
import matplotlib.pyplot as plt
labels = 'Frogs', 'hogs', 'dogs', 'logs'
sizes = [15,30,45,10]
explode = (0,0.1,0,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = '개구리', 'hogs', 'dogs', 'logs'
sizes = [15,30,45,10]
explode = (0,0.1,0,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```
